```python
#libraries
import pandas as pd
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
dataset=pd.read_csv('restaurants-challenge.csv', header=None)
#the dataset ranks Newyork restaurants according to their sanitary grade
dataset.head()
```

```python
dataset.shape
```

```python
transaction=[]
for i in range(0, 179731):

  transaction.append(
      [str(dataset.values[i, j])
  for j in range(0,3)]
  
  )
```

```python
transaction[0]
```

```python
#Encoding transactions
from mlxtend.preprocessing import TransactionEncoder
te= TransactionEncoder()
tranEncoded= te.fit_transform(transaction)
```

```python
tranEncoded
```

```python
te_trans=pd.DataFrame(tranEncoded, columns=te.columns_)
```

```python
te_trans.head()
```

#Association Rule

```python
model= apriori(te_trans, min_support=0.03, use_colnames=True)
```

```python
result= association_rules(model, metric='confidence', min_threshold=0.3)
```

```python
result.sort_values(by= 'lift', ascending=False)
```
