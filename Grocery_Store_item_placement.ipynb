```python
#libraries
import pandas as pd
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
data= pd.read_csv('groceries.csv')
```

```python
data.head()
```

```python
'''
The data above shows the number of items bought by each customer in a grocery store.
The first customer( customer 0), boought citrus fruit, semi-finished bread, margrine etc.
'''
```

```python
data.shape
```

```python
transaction=[]
for i in range(0, 9835):

  transaction.append(
      [str(data.values[i, j])
  for j in range(0,22)]
  
  )
```

```python
#Encoding transactions
from mlxtend.preprocessing import TransactionEncoder
```

```python
te= TransactionEncoder()
```

```python
tranEncoded= te.fit_transform(transaction)
```

```python
te_trans=pd.DataFrame(tranEncoded, columns=te.columns_)
```

```python
te_trans.sample(3)
```

```python
te_trans.drop(columns='nan', inplace=True)
```

```python
model= apriori(te_trans, min_support=0.03, use_colnames=True)
```

```python
result= association_rules(model, min_threshold=0.3)
result.sort_values(by='lift', ascending=False)
```
